#  Feature Descriptions

###  Demographic and customer information
- **gender**  Customers gender (Male/Female).
- **SeniorCitizen**  Whether the customer is a senior citizen (1 = yes, 0 = no).
- **Partner**  Whether the customer has a partner (Yes/No).
- **Dependents**  Whether the customer has dependents (children or others).

---

###  Customer relationship with the company
- **tenure**  Number of months the customer has been with the company.

---

###  Services subscribed
- **PhoneService**  Whether the customer has phone service.
- **MultipleLines**  Whether the customer has multiple phone lines.
- **OnlineSecurity**  Whether the customer has online security service.
- **OnlineBackup**  Whether the customer has online backup service.
- **DeviceProtection**  Whether the customer has device protection/insurance.
- **TechSupport**  Whether the customer has additional technical support.
- **StreamingTV**  Whether the customer has streaming TV service.
- **StreamingMovies**  Whether the customer has streaming movie service.
- **InternetService_Fiber optic**  Dummy variable: 1 if the customer uses fiber-optic internet.
- **InternetService_No**  Dummy variable: 1 if the customer has no internet service.

---

###  Billing information
- **PaperlessBilling**  Whether the customer uses paperless billing.
- **MonthlyCharges**  Monthly amount charged to the customer.
- **TotalCharges**  Total amount billed so far.

---

###  Contract and payment
- **Contract_One year**  Dummy variable: 1 if the customer has a one-year contract.
- **Contract_Two year**  Dummy variable: 1 if the customer has a two-year contract.
  - *(A monthly contract is implied when both dummies are 0.)*

- **PaymentMethod_Credit card (automatic)**  Payment via automatic credit card.
- **PaymentMethod_Electronic check**  Payment via electronic check.
- **PaymentMethod_Mailed check**  Payment via mailed paper check.

---

###  Target variable
- **Churn**  Whether the customer has cancelled their service (Yes/No).


**The Problem Statement**

The goal of this project is to predict whether a customer will churn (i.e., stop using the service) based on their historical data. The dataset provided includes customer demographic details, account information, and service usage patterns. By building a predictive model, the aim is to identify customers who are likely to churn, allowing the company to take proactive steps to retain them.

# 2. Basic cleaning

## 2.1 Clean TotalCharges (convert blanks to NaN, then to float)

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

## 2.2 Replace remaining NaN in TotalCharges

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

# 3. Keep correct data types

We keep everything as strings, except numeric fields like tenure, MonthlyCharges, TotalCharges.

```python
df['Churn'] = df['Churn'].astype(str)
df['gender'] = df['gender'].astype(str)
df['SeniorCitizen'] = df['SeniorCitizen'].apply(lambda x: "Yes" if x==1 else "No")
```

Note: SeniorCitizen = 0/1 is confusing.
Converting to "Yes"/"No" makes the dataset internally consistent (all categories are strings).

# 4. Create feature: tenure_group

```python
def bucket_tenure(t):
    if t < 12:
        return "0-12"
    elif t < 24:
        return "12-24"
    elif t < 48:
        return "24-48"
    else:
        return "48+"

df['tenure_group'] = df['tenure'].apply(bucket_tenure)
```

# 5. Remove obvious whitespace / normalize categories

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].str.strip()
```
